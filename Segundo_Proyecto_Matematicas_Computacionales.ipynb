```python
import numpy as np
import matplotlib.pyplot as plt


n = int(input("Ingrese el número de puntos (mínimo 5): "))
grado = int(input("Ingrese el grado del polinomio (mayor o igual a 2): "))


assert n >= 5, "El número de puntos debe ser al menos 5"
assert grado >= 2, "El grado del polinomio debe ser al menos 2"


print("Ingrese los puntos en el formato 'x y':")
puntos = [list(map(float, input().split())) for _ in range(n)]
x, y = zip(*puntos)


coeficientes = np.polyfit(x, y, grado)
polinomio = np.poly1d(coeficientes)


xp = np.linspace(min(x), max(x), 100)
plt.plot(x, y, 'o', xp, polinomio(xp), '-')
plt.title('Regresión polinomial')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
