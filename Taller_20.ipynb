```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr


x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([3.2, 0.4, -1, -1.4, -1.1, 0.6, 3.1])


coeficientes = np.polyfit(x, y, 2)
polinomio = np.poly1d(coeficientes)


y_ajuste = polinomio(x)
r = pearsonr(y, y_ajuste)[0]


plt.scatter(x, y, color='blue')
plt.plot(x, y_ajuste, color='red')
plt.show()


print(f"El polinomio de mejor ajuste es y = {coeficientes[0]}x^2 + {coeficientes[1]}x + {coeficientes[2]}")
print(f"Coeficiente de correlación (r): {r}")
```
