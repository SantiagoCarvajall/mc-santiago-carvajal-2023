```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([1, 3, 5, 7, 9, 11, 13, 15])
y = np.array([8.5, 13, 15, 16, 17, 17.5, 18, 19])

logx = np.log(x)
logy = np.log(y)
coeficientes_potencias = np.polyfit(logx, logy, 1)
polinomio_potencias = np.poly1d(coeficientes_potencias)
y_ajuste_potencias = np.exp(polinomio_potencias(logx))


coeficientes_crecimiento = np.polyfit(x, logy, 1)
polinomio_crecimiento = np.poly1d(coeficientes_crecimiento)
y_ajuste_crecimiento = np.exp(polinomio_crecimiento(x))


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue')
plt.plot(x, y_ajuste_potencias, color='red')
plt.title('Modelo de ecuación de potencias')

plt.subplot(1, 2, 2)
plt.scatter(x, y, color='blue')
plt.plot(x, y_ajuste_crecimiento, color='red')
plt.title('Modelo de razón de crecimiento')

plt.show()


print(f"El modelo de ecuación de potencias de mejor ajuste es y = {np.exp(coeficientes_potencias[1])} * x^{coeficientes_potencias[0]}")
print(f"El modelo de razón de crecimiento de mejor ajuste es y = {np.exp(coeficientes_crecimiento[1])} * e^({coeficientes_crecimiento[0]} * x)")
```
