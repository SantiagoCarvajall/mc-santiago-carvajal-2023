```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt


x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([2, 4, 5, 4, -3, 1, 12])


cs = CubicSpline(x, y)


estimacion = cs(4.75)


plt.scatter(x, y, color='blue')
plt.plot(x, cs(x), color='red')
plt.title('Trazadores cúbicos')
plt.show()


print(f"La estimación de f(4.75) con trazadores cúbicos es: {estimacion}")
```

```python
import numpy as np
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt


x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([2, 4, 5, 4, -3, 1, 12])


polinomio = lagrange(x, y)


estimacion = polinomio(4.75)


plt.scatter(x, y, color='blue')
plt.plot(x, polinomio(x), color='red')
plt.title('Interpolación de Lagrange')
plt.show()


print(f"El polinomio de interpolación de Lagrange es: {polinomio}")
print(f"La estimación de f(4.75) es: {estimacion}")
```
