```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress


x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([0.2, 0.5, 1.8, 3.4, 5.7, 9.0, 13.8])


coeficientes = np.polyfit(x, y, 1)
polinomio = np.poly1d(coeficientes)


y_ajuste = polinomio(x)
residuos = y - y_ajuste
sy = np.sqrt(np.sum(residuos**2) / (len(y) - 2))
syx = np.sqrt(np.sum(residuos**2) / len(y))
r = np.corrcoef(x, y)[0, 1]


plt.scatter(x, y, color='blue')
plt.plot(x, y_ajuste, color='red')
plt.show()


print(f"La línea de mejor ajuste es y = {coeficientes[0]}x + {coeficientes[1]}")
print(f"Desviación estándar (sy): {sy}")
print(f"Error estándar de la estimación (syx): {syx}")
print(f"Coeficiente de correlación (r): {r}")
```
