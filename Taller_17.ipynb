```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2.2, 3, 4.5, 6, 8.5, 12])


logy = np.log(y)


coeficientes = np.polyfit(x, logy, 1)
polinomio = np.poly1d(coeficientes)


y_ajuste = np.exp(polinomio(x))


plt.scatter(x, y, color='blue')
plt.plot(x, y_ajuste, color='red')
plt.show()


print(f"El modelo exponencial de mejor ajuste es y = exp({coeficientes[0]}x + {coeficientes[1]})")
```
